```python
#remove rows containing missing values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fill missing values with a specific value
df['year'].fillna(1000, inplace=True)
```

```python
df.head()
```

```python
# Fill missing values with the mean of the column
df.fillna(df.mean(), inplace=True)
```

```python
df.head(30)
```

```python
# Fill missing values with the median of the column
df['year'].fillna(df['year'].median(), inplace=True)
```

```python
df.head()
```

```python
#fill missing values with the previous or next non-missing value
# Forward fill
df.ffill(inplace=True)
df.head()
```

```python
# Backward fill
df.bfill(inplace=True)
df.head()
```

```python
# Linear interpolation (to estimate missing values based on the surrounding values)
df.interpolate(method='linear', inplace=True)
df.head()
```

```python
#drop columns with missing values
df.dropna(axis=1, inplace=True)
df.head()
```

```python
#Handling Categorical Data
#Replace missing values with the mode
df['seller_type'].fillna(df['seller_type'].mode()[0], inplace=True)
df.head(30)
```

```python
#Replace missing values with a specific category
df['transmission'].fillna('Automatic', inplace=True)
df.head(30)
```

```python
# Remove duplicates and update the existing DataFrame
df.drop_duplicates(inplace=True)
df.head(30)
```

```python
#Remove values beyond a certain threshold or percentile
upper_limit = df['selling_price'].quantile(0.75)
df_no_outliers = df[df['selling_price'] <= upper_limit]
print(df_no_outliers)
```

```python
# Convert a Series to integer
df['year'] = df['year'].astype(int)
df.head(10)
```

```python
# Convert a Series to float
df['selling_price'] = df['selling_price'].astype(float)
df.head(10)
```

```python
# Convert the entire DataFrame to a specific data type
df = df.astype({'selling_price': int, 'year': int, 'km_driven': float})
df.head(10)
```

```python
# Convert to float using eval()
df['km_driven'] = pd.eval(df['km_driven'].astype(float)).astype(str)
df.head(10)
```

```python
#The Z-score of a data point represents how many standard deviations it is from the mean
#method used to transform numerical data to have a standard normal distribution
#with a mean of 0 and a standard deviation of 1

from sklearn.preprocessing import StandardScaler

# values.reshape(-1,1) is used to convert 1D -> 2D
column_to_standardize = df['year'].values.reshape(-1, 1)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform the column
df['year_standardized'] = scaler.fit_transform(column_to_standardize)
df.head(10)
```

```python
# Manually standardize the column
df['selling_price_standardized'] = (df['selling_price'] - df['selling_price'].mean()) / df['selling_price'].std()
df.head(10)
```

```python
#The purpose of Min-Max scaling is to scale the data in a way
#that it falls within a pre-defined interval, usually [0, 1]
from sklearn.preprocessing import MinMaxScaler

column_to_normalize = df['year'].values.reshape(-1, 1)

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Fit and transform the column
df['year_normalized'] = scaler.fit_transform(column_to_normalize)
df.head(10)
```

```python
# Manually normalize the column
df['selling_price_normalized'] = (df['selling_price'] - df['selling_price'].min()) / (df['selling_price'].max() - df['selling_price'].min())
df.head(10)
```

```python
#The Robust Scaler is particularly useful when dealing with features that have outliers
#Unlike Min-Max scaling or Z-score normalization
#the Robust Scaler uses the median and the interquartile range (IQR) to scale the data
from sklearn.preprocessing import RobustScaler

column_to_scale = df['year'].values.reshape(-1, 1)

# Create a RobustScaler instance
scaler = RobustScaler()

# Fit and transform the column
df['year_scaled_robust'] = scaler.fit_transform(column_to_scale)
df.head(10)
```

```python
#The log1p function computes log(1+x), making it suitable for variables that may have zero values
df['selling_price'] = np.log1p(df['selling_price'])
df.head(10)
```

```python
df1=pd.read_csv('cars2.csv')
df1.head()
```

```python
#date parsing functions to convert string representations of dates into datetime objects
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Time'] = pd.to_datetime(df1['Time'], format='%H:%M:%S').dt.time
df1.head()
```

```python
# Extract Date Components
df1['day'] = df1['Date'].dt.day
df1['month'] = df1['Date'].dt.month
df1['year'] = df1['Date'].dt.year
df1['day_of_week'] = df1['Date'].dt.dayofweek
print(df1[['day', 'month', 'year', 'day_of_week']])
```

```python
# Extract Time Components
df1['hour'] = df1['Time'].dt.hour
df1['minute'] = df1['Time'].dt.minute
df1['second'] = df1['Time'].dt.second
print(df1[['hour', 'minute', 'second']])
```

```python
# Handling Time Zones
from pytz import timezone
df1['date_column_utc'] = df1['Time'].dt.tz_localize('UTC')
df1['date_column_est'] = df1['date_column_utc'].dt.tz_convert(timezone('US/Eastern'))
print(df1[['date_column_utc', 'date_column_est']])
```
